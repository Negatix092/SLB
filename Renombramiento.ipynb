```python
# codigo para listar los nombres de las carpetas en un directorio en orden alfabético 
import os
import glob

# Directorio donde se tienen las carpetas de interés

directorio_base = r'\\dir.slb.com\NSA\SAM_Collaborate\EC0037\00_Well_File\CONONACO'

# Función para listar los nombres de los archivos en la carpeta de interés
def listar_pozos(directorio):
    lista_pozos = []
    for pozo in os.listdir(directorio):
        ruta_pozo = os.path.join(directorio, pozo)
        if os.path.isdir(ruta_pozo):
            lista_pozos.append(pozo)
    return sorted(lista_pozos)

l = listar_pozos(directorio_base)

# Función para listar las carpetas en un directorio en orden alfabético

def listar(directorio):
    nombres_archivos = []
    for pozo in os.listdir(directorio):
        ruta = os.path.join(directorio, pozo)
        if os.path.isdir(ruta):
            # búsqueda recursiva en el subdirectorio
            nombre_archivo = listar(ruta)
            if nombre_archivo is not None:
                return nombre_archivo
        elif os.path.isfile(ruta):
            #retornar el nombre del primer archivo encontrado
            return os.path.basename(ruta)
    return None

for item in l:
    new_path = os.path.join(directorio_base, item)
    resultado = listar(new_path)
    archivo = resultado
    if resultado:
        print(f"Archivo encontrado para {item}: {archivo}")
    else:
        print(f"No se encontraron archivos para {item}")
```

```python
import pandas as pd
import os

# Path al archivo excel que se arma
excel_file_path = r'C:\Users\aperez143\OneDrive - SLB\Alex Perez\Reporte trimestral\Documentos\Shared Documents\New_Names.xlsx'
sheet_name = 'CULEBRA'

# Se lee el archivo excel
df = pd.read_excel(excel_file_path, sheet_name=sheet_name)
print(df)

# Path al directorio de interes 
directory_path = r'\\dir.slb.com\NSA\SAM_Collaborate\EC0037\00_Well_File\CULEBRA'

# Se ordenan las carpetas del directorio en orden alfabetico 
folders = sorted([f for f in os.listdir(directory_path) if os.path.isdir(os.path.join(directory_path, f))])

print(f"Number of folders: {len(folders)}")
print(f"Number of new names: {len(df)}")
# Check if the number of folder names matches the number of new names
if len(folders) != len(df):
    print("The number of folders and the number of new names do not match.")
else:
    for i, folder in enumerate(folders):
        new_folder_name = df.iloc[i, 0]  # Assuming the new names are in the first column
        original_folder_path = os.path.join(directory_path, folder)
        new_folder_path = os.path.join(directory_path, new_folder_name)

        if os.path.exists(original_folder_path):
            os.rename(original_folder_path, new_folder_path)
        else:
            print(f"Folder '{folder}' does not exist")

    print("Renaming complete.")
```
